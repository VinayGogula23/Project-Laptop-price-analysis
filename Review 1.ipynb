```python
lapdata['Display-inch']=lapdata['Display-inch'].str.replace('inch',"")
lapdata
```

```python
lapdata['Price-₹']=lapdata['Price-₹'].str.replace('₹',"").str.replace(',','')
lapdata
```

```python
Memory_Type=[]
for i in lapdata['Storage-TB']:
    Memory_Type.append(i.split()[2])
```

```python
lapdata['Memory_Type']=pd.Series(Memory_Type)
lapdata
```

```python
lapdata['Storage-TB']=lapdata['Storage-TB'].str.replace('SSD',"").str.replace('HDD',"").str.replace('EMMC',"")
```

```python
lapdata
```

```python
TG=[]
for i in lapdata['Storage-TB']:
    a=i.split()[1]
    b=i.split()[0]
    if a=='TB':
        TG.append(int(b))
    else:
        TG.append(int(b)/1024)
```

```python
lapdata['Storage-TB']=pd.Series(TG)
```
