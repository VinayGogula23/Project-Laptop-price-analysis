# DATA CLEANING

```python
lapdata.drop(labels='Unnamed: 0',axis=1,inplace=True)
lapdata
```

```python
lapdata.columns=['Brands', 'Operating_System','RAM', 'Memory-TB', 'Processor', 'bit', 'Generation',
       'Screen_Size_inch', 'Warranty', 'Price']
lapdata
```

```python
lapdata['Memory-TB']=lapdata['Memory-TB'].str.replace('[','').str.replace(']','').str.replace('\'','').str.replace(',','')
```

```python
lapdata
```

```python
lapdata['Screen_Size_inch']=lapdata['Screen_Size_inch'].str.replace('inch',"")
lapdata
```

```python
lapdata['Price']=lapdata['Price'].str.replace('₹',"").str.replace(',','')
lapdata
```

```python
Memory_Type=[]
for i in lapdata['Memory-TB']:
    Memory_Type.append(i.split()[2])
```

```python
lapdata['Memory_Type']=pd.Series(Memory_Type)
```

```python
lapdata['Memory-TB']=lapdata['Memory-TB'].str.replace('SSD',"").str.replace('HDD',"").str.replace('EMMC',"")
```

```python
lapdata
```

```python
TG=[]
for i in lapdata['Memory-TB']:
    a=i.split()[1]
    b=i.split()[0]
    if a=='TB':
        TG.append(int(b))
    else:
        TG.append(int(b)/1024)
```

```python
lapdata['Memory-TB']=pd.Series(TG)
```

```python
# import libraries
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
url='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=995dd11e-16dc-4b09-b635-c1685cf011ad&as-searchtext=lap&page='
rating=[]
x=[]
for j in range(1,35):
    url1=url+(str(j))
    page = requests.get(url1)
    soup=BeautifulSoup(page.text)
    print(j)
    print(url1)
    r=soup.find_all('div',class_="_3LWZlK")
    for i in r:
        rating.append(i.text)
    x.extend(rating[0:24])
```

```python
x2=pd.DataFrame(x)
x2
```
