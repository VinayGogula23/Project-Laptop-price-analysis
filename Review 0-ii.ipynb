```python
# import libraries
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
url='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=995dd11e-16dc-4b09-b635-c1685cf011ad&as-searchtext=lap&page='
Windows=[]
prices=[]
brand=[]
RAM=[]
SSD=[]
bit=[]
Screen_Size=[]
Warranty=[]
processor=[]
generation2=[]
rating=[]
for j in range(1,35):
    url1=url+(str(j))
    page = requests.get(url1)
    soup=BeautifulSoup(page.text)
    print(j)
    print(url1)
    b=soup.find_all('div',class_='_4rR01T')
    gen=soup.find_all('div',class_=('_3pLy-c row'))
    # Generations
    for i in gen:
        a=(re.findall(r'\d+th\sGen|\d{4}U|\d{4}H',i.text))
        if a:
            if len(a)>0:
                 generation2.append(a[0])
            else:
                 generation2.append(a)
        else:generation2.append(np.nan)
    
    b=soup.find_all('div',class_='_4rR01T')
    #Windows
    for i in b:
        a=re.findall(r'Windows\s\d+|Mac OS|DOS|\d+\sGB\sEMMC',i.text)
        #print(a)
        if len(a)>0: 
            Windows.append(a[0])
        else:
            Windows.append(np.nan)
    # Processor
    for i in b:
        a=re.findall(r'Core\si\d|Ryzen\s\d',i.text)
        if a:
            processor.append(a[0])
        else: 
            processor.append(np.nan)
     # Brand
    for i in b:
        brand.append(i.text.split()[0])
    
    # RAM
    for i in b:
        RAM.append(re.findall(r'\d+\sGB',i.text)[0].split()[0])
    
    # SSD
    for i in b:
        a=(re.findall(r'\d+\sGB\sSSD|\d+\sTB\sHDD|\d+\sTB\sSSD|\d+\sGB\sEMMC',i.text)[0].split())
        SSD.append(a)
    
    #Price
    p=soup.find_all('div',class_="_30jeq3 _1_WHN1")
    
    for i in p:
        prices.append(i.text)
    
    spacifications=soup.find_all('ul',class_='_1xgFaf')
    # bit
    
    for i in spacifications:
        a=re.findall(r'(\d+)\sbit',i.text)
        if len(a)>0:
            bit.append(a[0])
        else:
            bit.append(np.nan)
    
    # Screensize
    for i in spacifications:
        a=(re.findall(r'\d+\sinch',i.text))
        if a:
            Screen_Size.append(a[0])
        else:
            Screen_Size.append(np.nan)
    
    # Warranty
    for i in spacifications:
        a=re.findall(r'(\d)\s\w+\s\w+\sWarranty',i.text)
        if a:
            Warranty.append(a[0])
        else:
            Warranty.append(np.nan)
    
    
```
